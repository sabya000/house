```python
def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(convertRange)
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqr_feet)

        st = np.std(subdf.price_per_sqr_feet)

        gen_df = subdf[(subdf.price_per_sqr_feet > (m - st)) & (subdf.price_per_sqr_feet <= (m+st))]
        df_output = pd.concat([df_output, gen_df], ignore_index=True)
    return df_output
df = remove_outliers_sqft(df)
df.describe()
```

```python
def bhk_outlier_remove(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_status = {}
        for bhk, bhk_df in location_df.groupby('BHK'):
            bhk_status[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqr_feet),
                'std': np.std(bhk_df.price_per_sqr_feet),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('BHK'):
            stats = bhk_status.get(bhk-1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqr_feet<(stats['mean'])]. index.values)
    return df.drop(exclude_indices, axis='index')
```

```python
df = bhk_outlier_remove(df)

df.sample(5)
df.shape
```

```python
X = df.drop(columns=['price'])
y = df['price']

# X = df[['location', 'area', 'bedrooms', 'bathrooms']]
# y = df['price']
```

```python
from sklearn.linear_model import Lasso, Ridge
# from sklearn.preprocessing import OneHotEncoder, StandardScaler
# from sklearn.compose import make_column_transformer
# from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

```python
# column_trf = make_column_transformer(
#     (OneHotEncoder(sparse_output=False), ['location']),
#     remainder='passthrough'
# )

# sc = StandardScaler(with_mean=False)
# lr = LinearRegression()

# pipe = make_pipeline(column_trf, sc, lr)
# pipe.fit(X_train, y_train)

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression

# Categorical & numerical columns
# location	total_sqft	bath	BHK
categorical_cols = ['location']
numerical_cols = ['total_sqft', 'bath', 'BHK']

# Preprocessing
preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
    ('num', 'passthrough', numerical_cols)
])
```

```python
model_lr = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

model_lr.fit(X_train, y_train)
```

```python
y_pred_lr = model_lr.predict(X_test)
r2_score(y_test, y_pred_lr)
```

```python

model_ls = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', Lasso())
])

model_ls.fit(X_train, y_train)
```

```python
y_pred_ls = model_ls.predict(X_test)
r2_score(y_test, y_pred_ls)
```

```python
model_rg = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', Ridge())
])

model_rg.fit(X_train, y_train)
```

```python
y_pred_ri = model_rg.predict(X_test)
r2_score(y_test, y_pred_ri)
```
